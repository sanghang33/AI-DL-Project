```python
#   
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, RocCurveDisplay

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

#  
df = pd.read_csv("dataset/heart_disease.csv")   
print("Shape:", df.shape)
display(df.head())
df.info()

print("\nTarget distribution:")
display(df["HeartDisease"].value_counts())
display(df["HeartDisease"].value_counts(normalize=True))

# EDA -   ()
def compare_hist(col):
    plt.figure()
    df[df["HeartDisease"]==0][col].hist(alpha=0.6, bins=20)
    df[df["HeartDisease"]==1][col].hist(alpha=0.6, bins=20)
    plt.title(f"{col} distribution (0 vs 1)")
    plt.xlabel(col); plt.ylabel("count")
    plt.legend(["HeartDisease=0", "HeartDisease=1"])
    plt.show()

for col in ["Age", "MaxHR", "Oldpeak"]:
    compare_hist(col)

#EDA -     
plt.figure()
pd.crosstab(df["ChestPainType"], df["HeartDisease"], normalize="index").plot(kind="bar")
plt.title("HeartDisease rate by ChestPainType")
plt.xlabel("ChestPainType")
plt.ylabel("Rate")
plt.legend(title="HeartDisease")
plt.tight_layout()
plt.show()

# EDA -  
num_cols = ["Age", "RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Oldpeak", "HeartDisease"]
corr = df[num_cols].corr()

plt.figure(figsize=(8, 6))
im = plt.imshow(corr, aspect="auto")
plt.colorbar(im)
plt.xticks(range(len(num_cols)), num_cols, rotation=45, ha="right")
plt.yticks(range(len(num_cols)), num_cols)
plt.title("Correlation Heatmap (Numeric features)")
plt.tight_layout()
plt.show()

# /   
print("Cholesterol == 0 count:", (df["Cholesterol"]==0).sum())
print("RestingBP == 0 count:", (df["RestingBP"]==0).sum())

#  (0 -> NaN ->  ) +  +  + split
df2 = df.copy()

for col in ["Cholesterol", "RestingBP"]:
    df2[col] = df2[col].replace(0, np.nan)
    df2[col] = df2[col].fillna(df2[col].median())
    
    display(df2.head(10))      #     10 ()


y = df2["HeartDisease"].astype(int)
X = df2.drop(columns=["HeartDisease"])

X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

print("Train/Test shapes:", X_train_s.shape, X_test_s.shape)

# TOP15 Feature 

data_for_corr = X.copy()
data_for_corr["HeartDisease"] = y.values

corr_full = data_for_corr.corr(numeric_only=True)

top_features = (
    corr_full["HeartDisease"]
    .drop("HeartDisease")
    .abs()
    .sort_values(ascending=False)
    .head(15)
    .index
    .tolist()
)

cols = top_features + ["HeartDisease"]
corr_top = corr_full.loc[cols, cols]

plt.figure(figsize=(10, 7))
im = plt.imshow(corr_top, aspect="auto")
plt.colorbar(im)
plt.xticks(range(len(cols)), cols, rotation=60, ha="right")
plt.yticks(range(len(cols)), cols)
plt.title("Correlation Heatmap (Top 15 features vs HeartDisease)")
plt.tight_layout()
plt.show()

#   
model = keras.Sequential([
    layers.Input(shape=(X_train_s.shape[1],)),
    layers.Dense(64, activation="relu"),
    layers.Dropout(0.30),
    layers.Dense(32, activation="relu"),
    layers.Dropout(0.20),
    layers.Dense(1, activation="sigmoid"),
])

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=["accuracy", keras.metrics.AUC(name="auc")]
)

model.summary()

# 
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_auc", mode="max", patience=15, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor="val_auc", mode="max", patience=7, factor=0.5),
]

history = model.fit(
    X_train_s, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

plt.figure()
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss")
plt.xlabel("epoch"); plt.ylabel("loss")
plt.legend(["train", "val"])
plt.show()

plt.figure()
plt.plot(history.history["auc"])
plt.plot(history.history["val_auc"])
plt.title("AUC")
plt.xlabel("epoch"); plt.ylabel("auc")
plt.legend(["train", "val"])
plt.show()

# 
proba = model.predict(X_test_s).ravel()
pred = (proba >= 0.5).astype(int)

auc = roc_auc_score(y_test, proba)
cm = confusion_matrix(y_test, pred)
report = classification_report(y_test, pred, digits=4)

print("\nROC-AUC:", round(auc, 4))
print("\nConfusion Matrix:\n", cm)
plt.figure(figsize=(4, 3))
plt.imshow(cm, aspect="auto")
plt.colorbar()
plt.xticks([0, 1], ["Pred 0", "Pred 1"])
plt.yticks([0, 1], ["True 0", "True 1"])
plt.title("Confusion Matrix")

#    
for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.tight_layout()
plt.show()

print("\nClassification Report:\n", report)

plt.figure()
RocCurveDisplay.from_predictions(y_test, proba)
plt.title("ROC Curve")
plt.show()
```
